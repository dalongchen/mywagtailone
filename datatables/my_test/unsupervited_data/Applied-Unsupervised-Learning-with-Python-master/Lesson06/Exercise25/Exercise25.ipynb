# Exercise 2 - TSNE MNIST - Perplexity

In this exercise we will use the MNIST dataset (provided in the accompanying source code) to explore the effect of perplexity on the t-SNE manifold.

For this exercise import:
* pickle
* numpy
* PCA from scikit-learn
* TSNE from scikit-learn
* matplotlib

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

Load the MNIST dataset

```python
with open('mnist.pkl', 'rb') as f:
    mnist = pickle.load(f)
```

In this exercise we are going to use PCA on the dataset to reduce extract only the first 30 components.

```python
model_pca = PCA(n_components=30)
mnist_pca = model_pca.fit_transform(mnist['images'].reshape((-1, 28 ** 2)))
```

In this exercise we are investigating the effect of perplexity on the t-SNE manifold.  Iterate through a model / plot look with a perplexity of 3, 30 and 300.

```python
MARKER = ['o', 'v', '1', 'p' ,'*', '+', 'x', 'd', '4', '.']
for perp in [3, 30, 300]:
    model_tsne = TSNE(random_state=0, verbose=1, perplexity=perp)
    mnist_tsne = model_tsne.fit_transform(mnist_pca)

    plt.figure(figsize=(10, 7))
    plt.title(f'Low Dimensional Representation of MNIST (perplexity = {perp})');
    for i in range(10):
        selections = mnist_tsne[mnist['labels'] == i]
        plt.scatter(selections[:,0], selections[:,1], alpha=0.2, marker=MARKER[i], s=5);
        x, y = selections.mean(axis=0)
        plt.text(x, y, str(i), fontdict={'weight': 'bold', 'size': 30})    
```
