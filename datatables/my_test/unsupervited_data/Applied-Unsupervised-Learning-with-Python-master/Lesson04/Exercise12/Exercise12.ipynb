```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris-data.csv')
df.head()
```

```python
df = df[['Sepal Length', 'Sepal Width']]
df.head()
```

```python
eigenvectors, eigenvalues, _ = np.linalg.svd(df.values, full_matrices=False)
```

```python
eigenvalues
```

```python
eigenvalues = np.cumsum(eigenvalues)
eigenvalues
```

```python
eigenvalues /= eigenvalues.max()
eigenvalues
```
